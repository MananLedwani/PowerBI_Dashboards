```python
sns.countplot(x = 'Sex', data = df)
#There are more males than females in the ship
```

```python
df.groupby('Sex').size()
```

```python
sns.distplot(df['Age'])
```

```python
df.groupby(['Survived','Sex']).size()
#195 females and 93 males survived
#64 females and 360 males died
```

```python
ax = sns.countplot(data = df,x = df['Survived'],hue = df['Sex'])
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Pclass','Sex']).size()
'''
There are 83 females and 101 males in passenger class 1
There are 74 females and 99 males in passenger class 2
There are 102 females and 253 males in passenger class 3
'''
```

```python
ax1 = sns.countplot(data = df,x = df['Pclass'],hue = df['Sex'])
for bars in ax1.containers:
  ax1.bar_label(bars)
```

```python
ax1 = sns.countplot(data = df,x = df['Pclass'],hue = df['Survived'])
for bars in ax1.containers:
  ax1.bar_label(bars)
'''
64 people from passenger class 1 didn't survive and 120 survived.
90 people from passenger class 1 didn't survive and 83 survived.
270 people from passenger class 1 didn't survive and 85 survived.
'''
```
